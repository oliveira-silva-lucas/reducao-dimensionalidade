```python
def converter_tons_cinza(imagem):
    largura, altura = imagem.size
    imagem_tons_cinza = Image.new("L", (largura, altura)) 

    for y in range(altura):
        for x in range(largura):
            r, g, b = imagem.getpixel((x, y))
            valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
            imagem_tons_cinza.putpixel((x, y), valor_cinza)

    return imagem_tons_cinza


def binarizar_imagem_tons_cinza(imagem_cinza, limiar=127):
    largura, altura = imagem_cinza.size
    imagem_binaria = Image.new("1", (largura, altura)) 

    for y in range(altura):
        for x in range(largura):
            valor_cinza = imagem_cinza.getpixel((x, y))
            #limiar é o ponto/valor que decide se irá virar cinza ou branco
            valor_binario = 1 if valor_cinza > limiar else 0
            imagem_binaria.putpixel((x, y), valor_binario)

    return imagem_binaria
```

```python
from PIL import Image
from IPython.display import display

imagem =Image.open("mandela.jpg")
display(imagem)

tons_cinza = converter_tons_cinza(imagem)
display(tons_cinza)

binarizada = binarizar_imagem_tons_cinza(tons_cinza)
display(binarizada,185)
```
